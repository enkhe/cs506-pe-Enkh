# Part I.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageFilter, ImageEnhance

# Assuming you are in a Jupyter Notebook environment for this magic command
# %matplotlib inline

img = Image.open('PE05-input.jpg')

# Different ways to apply image filters
img = img.filter(ImageFilter.DETAIL)
img = img.filter(ImageFilter.SHARPEN)
img = img.filter(ImageFilter.MedianFilter)
img = img.filter(ImageFilter.UnsharpMask(5.0, 200, 20))

# Histogram equalization from ImageOps
img = ImageOps.equalize(img)

# Using Enhancer
img = ImageEnhance.Sharpness(img).enhance(-2.0)

plt.imshow(img)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageFilter, ImageEnhance

def process_and_display_image(image_path):
    # Open the original image
    img = Image.open(image_path)

    # Create a figure to display the quadrants
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # First quadrant: Original image
    axes[0, 0].imshow(img)
    axes[0, 0].set_title("Original Image")
    axes[0, 0].axis('on')

    # Second quadrant: Histogram equalized image
    img = ImageOps.equalize(img)
    axes[0, 1].imshow(img)
    axes[0, 1].set_title("Histogram Equalized")
    axes[0, 1].axis('on')

    # Third quadrant: Brightened image
    img = ImageEnhance.Brightness(img).enhance(1.5)
    axes[1, 0].imshow(img)
    axes[1, 0].set_title("Brightened Image")
    axes[1, 0].axis('on')

    # Fourth quadrant: Unsharp mask applied
    img = img.filter(ImageFilter.UnsharpMask(5.0, 200, 20))
    axes[1, 1].imshow(img)
    axes[1, 1].set_title("Unsharp Mask")
    axes[1, 1].axis('on')

    # Set custom x-axis and y-axis ticks for each subplot
    for ax in axes.flatten():
        #.set_xticks([0, 100, 200, 300, 400, 500])
        #ax.set_yticks([300, 200, 100])
        ax.set_xlim(0, 500)
        ax.set_ylim(400, 100)
        
    plt.tight_layout()
    plt.show()

# Example usage
process_and_display_image('PE05-input.jpg')

```

# Part II.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageFilter, ImageEnhance

def process_and_display_image(image_path):
    """
    This function processes an image and displays it in 4 quadrants:
    1. Original Image
    2. Histogram Equalized Image
    3. Brightened Image
    4. Image with Unsharp Mask
    """
    # Open the original image
    img = Image.open(image_path)

    # Create a figure to display the quadrants
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # First quadrant: Original image
    axes[0, 0].imshow(img)
    axes[0, 0].set_title("Original Image")
    axes[0, 0].axis('off')

    # Second quadrant: Histogram equalized image
    hist_eq_img = ImageOps.equalize(img)
    axes[0, 1].imshow(hist_eq_img)
    axes[0, 1].set_title("Histogram Equalized")
    axes[0, 1].axis('off')

    # Third quadrant: Brightened image
    bright_img = ImageEnhance.Brightness(img).enhance(4.0)
    axes[1, 0].imshow(bright_img)
    axes[1, 0].set_title("Brightened Image")
    axes[1, 0].axis('off')

    # Fourth quadrant: Unsharp mask applied
    unsharp_img = img.filter(ImageFilter.UnsharpMask(5.0, 200, 20))
    axes[1, 1].imshow(unsharp_img)
    axes[1, 1].set_title("Unsharp Mask")
    axes[1, 1].axis('off')

    # Display the figure
    plt.tight_layout()
    plt.show()

# Example usage
process_and_display_image('PE05-input.jpg')
```
